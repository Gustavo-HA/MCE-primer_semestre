## Herencia

```python
class Animal:
    def __init__(self, nombre):
        self.nombre = nombre

    def hacer_sonido(self):
        pass

class Perro(Animal):
    def hacer_sonido(self):
        return "Woof!"

class Ave(Animal):
    def hacer_sonido(self):
        return "Canto melodías"
```

```python
class Animal:
    def __init__(self, nombre, tipo):
        self.__nombre = nombre
        self.__tipo = tipo

    def emitir_sonido(self):
        pass

    def get_nombre(self):
        return self.__nombre

    def set_nombre(self, nuevo_nombre):
        self.__nombre = nuevo_nombre

    def get_tipo(self):
        return self.__tipo

    def set_tipo(self, nuevo_tipo):
        self.__tipo = nuevo_tipo

    def __str__(self):
        return f"{self.__nombre} ({self.__tipo})"

    def __eq__(self, otro):
        return self.__nombre == otro.__nombre and self.__tipo == otro.__tipo

    def __ne__(self, otro):
        return not self.__eq__(otro)

    def __lt__(self, otro):
        return self.__nombre < otro.__nombre

    def __le__(self, otro):
        return self.__nombre <= otro.__nombre

    def __gt__(self, otro):
        return self.__nombre > otro.__nombre

    def __ge__(self, otro):
        return self.__nombre >= otro.__nombre

    def __len__(self):
        return len(self.__nombre)

    def __del__(self):
        print(f"¡Adiós, {self.__nombre}!")

class Perro(Animal):
    def emitir_sonido(self):
        return "¡Guau!"

class Gato(Animal):
    def emitir_sonido(self):
        return "¡Miau!"

class Pajaro(Animal):
    def emitir_sonido(self):
        return "¡Pío pío!"

class Perro_Militar(Perro):
  def __init__(self, nombre, tipo):
      super().__init__(nombre, tipo);

  def emitir_sonido(self):
      return super().emitir_sonido() + "militar";

class Vehiculo:
    def __init__(self, marca, modelo):
        self.__marca = marca
        self.__modelo = modelo

    def conducir(self):
        pass

    def get_marca(self):
        return self.__marca

    def set_marca(self, nueva_marca):
        self.__marca = nueva_marca

    def get_modelo(self):
        return self.__modelo

    def set_modelo(self, nuevo_modelo):
        self.__modelo = nuevo_modelo

    def __str__(self):
        return f"{self.__marca} {self.__modelo}"

    def __eq__(self, otro):
        return self.__marca == otro.__marca and self.__modelo == otro.__modelo

    def __ne__(self, otro):
        return not self.__eq__(otro)

    def __del__(self):
        print(f"¡Hasta luego, {self.__marca} {self.__modelo}!")

class Coche(Vehiculo):
    def conducir(self):
        return "El coche está en movimiento"

class Motocicleta(Vehiculo):
    def conducir(self):
        return "La motocicleta está en movimiento"


class Pago_Impuestos():
  def __init__(self, registro):
      self.registro = registro;

  def set_registro(self, registro):
    self.registro = registro

  def pagar(self, dinero):
    print(f"La unidad con registro {self.registro} ha pagado la cantidad de {dinero}");


class Coche_Responsable(Coche, Pago_Impuestos):
  def __init__(self, marca, modelo, registro):
      super().__init__(marca, modelo);
      self.set_registro(registro);





class Utilidades:
    @staticmethod
    def mostrar_mensaje(mensaje):
        print(f"Mensaje: {mensaje}")

def interactuar(animal, vehiculo):
    print(f"{animal} interactúa con {vehiculo}:")
    print(f"{animal} dice:", animal.emitir_sonido())
    print(f"{vehiculo}:", vehiculo.conducir())
    Utilidades.mostrar_mensaje("Interacción completada.")
    print()

def main():
    perro = Perro("Max", "Canino")
    gato = Gato("Luna", "Felino")
    coche = Coche("Toyota", "Corolla")
    moto = Motocicleta("Harley-Davidson", "Sportster")

    interactuar(perro, coche)
    interactuar(gato, moto)

    print("Marca de la moto:", moto.get_marca())
    moto.set_marca("Ducati")
    print("Nueva marca de la moto:", moto.get_marca())


    perro_militar = Perro_Militar("Frida", "Busqueda");
    coche_con_registro = Coche_Responsable("VW", "Jetta", "XXX");

    interactuar(perro_militar, coche_con_registro);

if __name__ == "__main__":
    main()
```
