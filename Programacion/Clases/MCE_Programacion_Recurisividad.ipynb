# Problema: Suma de Dgitos Recursiva

Dado un nmero entero positivo, tu tarea es calcular la suma de sus dgitos de manera recursiva.

Ejemplo:

Entrada: 12345
Salida: 15 (1 + 2 + 3 + 4 + 5)


```python
def suma_digitos_recursiva(n):
    if n == 0:
        return 0
    else:
        return n % 10 + suma_digitos_recursiva(n // 10)

# Prueba la funcin
numero = 12345
resultado = suma_digitos_recursiva(numero)
print(resultado)
```

#Algoritmo de Euclides

El algoritmo de Euclides es un mtodo eficiente y antiguo para encontrar el mximo comn divisor (MCD) de dos nmeros enteros. El MCD de dos nmeros es el mayor nmero que los divide a ambos sin dejar un residuo.

#### Descripcin del Algoritmo:

1. **Paso 1: Comparacin de Nmeros**  
   Dados dos nmeros enteros \(a\) y \(b\) (donde \(a >= b\)), si \(b = 0\), entonces el MCD es \(a\).

2. **Paso 2: Divisin y Residuo**  
   Si \(b != 0\), divide \(a\) entre \(b\) y calcula el residuo \(r\), es decir, \(r = a \mod b\).

3. **Paso 3: Recursin**  
   Reemplaza \(a\) por \(b\) y \(b\) por \(r\). Luego repite el proceso con los nuevos valores de \(a\) y \(b\).

4. **Paso 4: Terminacin**  
   El proceso contina iterativamente hasta que el residuo \(r\) sea 0. En ese punto, el valor de \(b\) es el MCD de los dos nmeros originales.

#### Ejemplo:

Calculemos el MCD de 56 y 42 usando el algoritmo de Euclides:

1. \(a = 56\), \(b = 42\)
2. \(56 \mod 42 = 14\) (entonces \(r = 14\))
3. Ahora, reemplazamos \(a\) por 42 y \(b\) por 14:
   - \(a = 42\), \(b = 14\)
4. \(42 \mod 14 = 0\) (entonces \(r = 0\))

Aqu el residuo es 0, as que el MCD es 14.



```python
def gcd(a, b):
    # Caso base: si b es 0, el MCD es a
    if b == 0:
        return a
    # Llamada recursiva: gcd(b, residuo de a/b)
    else:
        return gcd(b, a % b)

# Ejemplo de uso
a = 56
b = 42
resultado = gcd(a,b)
print(f"El MCD de {a} y {b} es {resultado}")
```

# Problema: Torres de Hanoi

Las Torres de Hanoi es un rompecabezas matemtico clsico que consiste en tres varillas y una serie de discos de diferentes tamaos que pueden deslizarse sobre cualquier varilla. El rompecabezas comienza con los discos ordenados en una varilla en orden ascendente de tamao, el objetivo es mover la pila completa a otra varilla respetando las siguientes reglas:

Solo se puede mover un disco a la vez.
Un disco ms grande nunca puede colocarse sobre uno ms pequeo.
Dado el nmero de discos y las tres varillas (denominadas A, B y C), tu tarea es imprimir los pasos necesarios para mover todos los discos de la varilla A a la varilla C utilizando la recursividad.

```python
def torres_hanoi(n, origen, auxiliar, destino):
    if n == 1:
        print(f"Mueve disco 1 de {origen} a {destino}")
        return
    torres_hanoi(n-1, origen, destino, auxiliar)
    print(f"Mueve disco {n} de {origen} a {destino}")
    torres_hanoi(n-1, auxiliar, origen, destino)
# Prueba la funcin
num_discos = 6
torres_hanoi(num_discos, 'A', 'B', 'C')
```

# Problema: Laberinto con Obstculos

Dado un laberinto rectangular representado como una matriz booleana, donde las casillas vacas se representan como True y las casillas bloqueadas como False, tu tarea es encontrar un camino desde la casilla de inicio (esquina superior izquierda) hasta la casilla de destino (esquina inferior derecha), evitando los obstculos.

Implementa una funcin recursiva en un lenguaje de programacin de tu eleccin que resuelva este problema. Debes encontrar el camino y devolver una lista de coordenadas que representen el camino desde la casilla de inicio hasta la casilla de destino. Si no hay un camino posible, devuelve una lista vaca.

```python
def encontrar_camino_lab(laberinto, x, y):
    if x < 0 or x >= len(laberinto) or y < 0 or y >= len(laberinto[0]) or not laberinto[x][y]:
        return []

    if x == len(laberinto) - 1 and y == len(laberinto[0]) - 1:
        return [(x, y)]

    laberinto[x][y] = False  # Marcar la casilla actual como visitada

    # Explorar las cuatro direcciones posibles
    for dx, dy in [(1, 0), (-1, 0), (0, 1), (0, -1)]:
        camino = encontrar_camino_lab(laberinto, x + dx, y + dy)
        if camino:
            return [(x, y)] + camino

    return []

# Prueba la funcin
laberinto = [
    [True, True, False, True],
    [False, True, True, False],
    [True, False, True, True],
    [True, True, True, True]
]

inicio = (0, 0)
fin = (3, 3)
camino = encontrar_camino_lab(laberinto, inicio[0], inicio[1])
if camino:
    print("Camino encontrado:", camino)
else:
    print("No se encontr un camino.")
```
