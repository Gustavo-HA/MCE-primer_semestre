En programación, la **recusrsión** es un método de simplificación de un
problema complejo en la división de este en subproblemas del mismo tipo. Esta técnica de programación se conoce como divide y vencerás y es el núcleo en el diseño de numerosos algoritmos de gran importancia, así como también es parte fundamental de la programación dinámica.


---



Es decir, la recursión es una técnica que consistente en definir una función en términos de sí misma.



---


Se deben satisfacer dos condiciones para que se pueda resolver un problema recursivamente:



	Primera: El problema se debe escribir en forma recursiva.


	Segunda: La sentencia del problema debe incluir una condición de fin.



Uno de las funciones más famosas que se puede hacer es la de la sumatoria

```python
def sumatoria(n):
  if n<=0:# Condición de fin o paro
    return 0;
  return n + sumatoria(n-1) # llamada recursiva

print(sumatoria(n=0))
```

La función de fibonacci también esta definida en relación a sí misma.

Se define como f(n) = f(n-1) + f(n-2), y f(1) = 1 y f(2) = 1

```python
def fibonacci(n):
  if n<=2:
    return 1;
  return fibonacci(n-1) + fibonacci(n-2)

print(fibonacci(3))
```

# Problema: Suma de Dígitos Recursiva

Dado un número entero positivo, tu tarea es calcular la suma de sus dígitos de manera recursiva.

Ejemplo:

Entrada: 12345
Salida: 15 (1 + 2 + 3 + 4 + 5)


```python
def suma_digitos_recursiva(n):
    if n == 0:
        return 0
    else:
        return n % 10 + suma_digitos_recursiva(n // 10)

# Prueba la función
numero = 12345
resultado = suma_digitos_recursiva(numero)
print(resultado)
```

#Algoritmo de Euclides

El algoritmo de Euclides es un método eficiente y antiguo para encontrar el máximo común divisor (MCD) de dos números enteros. El MCD de dos números es el mayor número que los divide a ambos sin dejar un residuo.

#### Descripción del Algoritmo:

1. **Paso 1: Comparación de Números**  
   Dados dos números enteros \(a\) y \(b\) (donde \(a >= b\)), si \(b = 0\), entonces el MCD es \(a\).

2. **Paso 2: División y Residuo**  
   Si \(b != 0\), divide \(a\) entre \(b\) y calcula el residuo \(r\), es decir, \(r = a \mod b\).

3. **Paso 3: Recursión**  
   Reemplaza \(a\) por \(b\) y \(b\) por \(r\). Luego repite el proceso con los nuevos valores de \(a\) y \(b\).

4. **Paso 4: Terminación**  
   El proceso continúa iterativamente hasta que el residuo \(r\) sea 0. En ese punto, el valor de \(b\) es el MCD de los dos números originales.

#### Ejemplo:

Calculemos el MCD de 56 y 42 usando el algoritmo de Euclides:

1. \(a = 56\), \(b = 42\)
2. \(56 \mod 42 = 14\) (entonces \(r = 14\))
3. Ahora, reemplazamos \(a\) por 42 y \(b\) por 14:
   - \(a = 42\), \(b = 14\)
4. \(42 \mod 14 = 0\) (entonces \(r = 0\))

Aquí el residuo es 0, así que el MCD es 14.



```python
def gcd(a, b):
    # Caso base: si b es 0, el MCD es a
    if b == 0:
        return a
    # Llamada recursiva: gcd(b, residuo de a/b)
    else:
        return gcd(b, a % b)

# Ejemplo de uso
a = 56
b = 42
resultado = gcd(a,b)
print(f"El MCD de {a} y {b} es {resultado}")
```

# Problema: Torres de Hanoi

Las Torres de Hanoi es un rompecabezas matemático clásico que consiste en tres varillas y una serie de discos de diferentes tamaños que pueden deslizarse sobre cualquier varilla. El rompecabezas comienza con los discos ordenados en una varilla en orden ascendente de tamaño, el objetivo es mover la pila completa a otra varilla respetando las siguientes reglas:

Solo se puede mover un disco a la vez.
Un disco más grande nunca puede colocarse sobre uno más pequeño.
Dado el número de discos y las tres varillas (denominadas A, B y C), tu tarea es imprimir los pasos necesarios para mover todos los discos de la varilla A a la varilla C utilizando la recursividad.

```python
def torres_hanoi(n, origen, auxiliar, destino):
    if n == 1:
        print(f"Mueve disco 1 de {origen} a {destino}")
        return
    torres_hanoi(n-1, origen, destino, auxiliar)
    print(f"Mueve disco {n} de {origen} a {destino}")
    torres_hanoi(n-1, auxiliar, origen, destino)
# Prueba la función
num_discos = 6
torres_hanoi(num_discos, 'A', 'B', 'C')
```

# Problema: Laberinto con Obstáculos

Dado un laberinto rectangular representado como una matriz booleana, donde las casillas vacías se representan como True y las casillas bloqueadas como False, tu tarea es encontrar un camino desde la casilla de inicio (esquina superior izquierda) hasta la casilla de destino (esquina inferior derecha), evitando los obstáculos.

Implementa una función recursiva en un lenguaje de programación de tu elección que resuelva este problema. Debes encontrar el camino y devolver una lista de coordenadas que representen el camino desde la casilla de inicio hasta la casilla de destino. Si no hay un camino posible, devuelve una lista vacía.

```python
def encontrar_camino_lab(laberinto, x, y):
    if x < 0 or x >= len(laberinto) or y < 0 or y >= len(laberinto[0]) or not laberinto[x][y]:
        return []

    if x == len(laberinto) - 1 and y == len(laberinto[0]) - 1:
        return [(x, y)]

    laberinto[x][y] = False  # Marcar la casilla actual como visitada

    # Explorar las cuatro direcciones posibles
    for dx, dy in [(1, 0), (-1, 0), (0, 1), (0, -1)]:
        camino = encontrar_camino_lab(laberinto, x + dx, y + dy)
        if camino:
            return [(x, y)] + camino

    return []

# Prueba la función
laberinto = [
    [True, True, False, True],
    [False, True, True, False],
    [True, False, True, True],
    [True, True, True, True]
]

inicio = (0, 0)
fin = (3, 3)
camino = encontrar_camino_lab(laberinto, inicio[0], inicio[1])
if camino:
    print("Camino encontrado:", camino)
else:
    print("No se encontró un camino.")
```
