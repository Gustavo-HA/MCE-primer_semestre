# Ejercicio

La sumatoria de los primeros N números enteros se define como 1+2+3+...+N.

Eso quiere decir que sumatoria(1) = 1, sumatoria(2) = 3, sumatoria (3) = 6, etc.

Programa una función que calcule la sumatoria de las sumatorias hasta N, por ejemplo, si N = 3, entonces el resultado sería sumatoria(1) + sumatoria(2) + sumatoria(3), lo cual equivale a 1+3+6, es decir, 10.

```python
def sumatoriaDeSumatorias(n=3):
  res = 0;
  res = (n*n*n + 3*n*n + 2*n)//6


  return res



print(sumatoriaDeSumatorias(3))
```
